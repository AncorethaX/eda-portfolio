# 📊 Análisis Exploratorio de Datos (EDA)
Este notebook demuestra un EDA básico con un dataset simulado.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear dataset simulado
df = pd.DataFrame({
    'edad': [23, 45, 31, 35, 29, 41, 50, 38, 27, 33],
    'ingreso': [500, 1200, 800, 950, 700, 1100, 1500, 1000, 650, 900]
})
df.head()
```

## Descripción del dataset

```python
df.describe()
```

## Histograma de edades

```python
df['edad'].plot(kind='hist')
plt.title('Histograma de edades')
plt.xlabel('Edad')
plt.show()
```
